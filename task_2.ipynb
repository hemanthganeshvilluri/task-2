```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
hg=sns.load_dataset('titanic')
print(hg.head())
print(hg.info())
```

```python
print(hg.describe())
```

```python
plt.figure(figsize=(6,6))
plt.hist(hg['age'],bins=10,color='orange',edgecolor='black')
plt.title("histogram for age")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```

```python
sns.boxplot(data=hg['fare'],color='red')
plt.title("outliers")
plt.show()
```

```python
sns.pairplot(hg.iloc[:,:4])
plt.show()
```

```python
correlation=hg.corr(numeric_only=True)
plt.figure(figsize=(6,6))
sns.heatmap(correlation,annot=True,cmap='Blues')
plt.title("correlation values")
plt.show()
```

```python
plt.plot(hg['pclass'],hg['fare'])
plt.title("fare trend")
plt.xlabel("fare")
plt.ylabel("pclass")
plt.show()
```

```python
q1=hg['fare'].quantile(0.25)
q3=hg['fare'].quantile(0.75)
iqr=q3-q1
hg=hg[(hg['fare']>=q1-1.5*iqr)&(hg['fare']<=q3+1.5*iqr)]
print("outliers found")
print(hg['fare'])
```

```python
print(hg.shape)
```

```python
gh=px.scatter(hg,x='age',y='fare',color='sex')
gh.show()
```

```python
gh=px.histogram(hg,x='age',y='fare',nbins=10)
gh.show()
```
